```python
import requests
from textblob import TextBlob
from datetime import datetime

# Fetch real live news using GNews API (free, no key needed)
url = "https://gnews.io/api/v4/top-headlines?category=technology&lang=en&max=10&apikey=demo"

print("⏳ Fetching live tech news...")
response = requests.get(url)
data = response.json()

articles = data.get('articles', [])

if not articles:
    # Backup news if API limit reached
    articles = [
        {"title": "Artificial Intelligence is transforming the healthcare industry globally", "source": {"name": "TechCrunch"}},
        {"title": "Japan leads the world in robotics and automation technology", "source": {"name": "Forbes"}},
        {"title": "Cloud computing market grows rapidly across Asia Pacific region", "source": {"name": "Reuters"}},
        {"title": "Python remains the most popular programming language in 2024", "source": {"name": "StackOverflow"}},
        {"title": "Electric vehicles face supply chain challenges worldwide", "source": {"name": "Bloomberg"}},
        {"title": "Cybersecurity threats increase as remote work expands globally", "source": {"name": "BBC"}},
        {"title": "Semiconductor shortage impacts global technology production", "source": {"name": "CNBC"}},
        {"title": "Renewable energy investments reach record highs this year", "source": {"name": "Guardian"}},
        {"title": "Space exploration companies compete for lunar mission contracts", "source": {"name": "NASA"}},
        {"title": "Quantum computing breakthrough achieved by research team", "source": {"name": "MIT"}}
    ]
    print("📡 Using sample news dataset...")

print(f"✅ {len(articles)} articles fetched!\n")

# Analyze sentiment for each headline
print("🤖 AI Sentiment Analysis Results:")
print("=" * 60)

results = []
for i, article in enumerate(articles, 1):
    title = article['title']
    source = article['source']['name']

    # AI Sentiment Analysis
    blob = TextBlob(title)
    polarity = blob.sentiment.polarity

    if polarity > 0.1:
        sentiment = "🟢 POSITIVE"
    elif polarity < -0.1:
        sentiment = "🔴 NEGATIVE"
    else:
        sentiment = "🟡 NEUTRAL"

    results.append({
        "title": title,
        "source": source,
        "sentiment": sentiment,
        "score": round(polarity, 2)
    })

    print(f"{i}. [{sentiment}] (Score: {polarity:.2f})")
    print(f"   📰 {title}")
    print(f"   Source: {source}\n")

print("=" * 60)
print("✅ Sentiment Analysis Complete!")
```

```python
# Generate automated summary report
print("⏳ Generating report...")

positive = [r for r in results if "POSITIVE" in r['sentiment']]
negative = [r for r in results if "NEGATIVE" in r['sentiment']]
neutral = [r for r in results if "NEUTRAL" in r['sentiment']]

report = f"""
╔══════════════════════════════════════════════════════════╗
       🤖 AI-POWERED NEWS SENTIMENT ANALYSIS REPORT
╚══════════════════════════════════════════════════════════╝

📅 Generated: {datetime.now().strftime("%Y-%m-%d %H:%M:%S")}
📊 Total Articles Analyzed: {len(results)}

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
📈 SUMMARY
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
🟢 Positive News : {len(positive)} articles ({len(positive)/len(results)*100:.0f}%)
🔴 Negative News : {len(negative)} articles ({len(negative)/len(results)*100:.0f}%)
🟡 Neutral News  : {len(neutral)} articles ({len(neutral)/len(results)*100:.0f}%)

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
🏆 TOP POSITIVE HEADLINE
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
{positive[0]['title'] if positive else 'None'}

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
⚠️  TOP NEGATIVE HEADLINE
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
{negative[0]['title'] if negative else 'None'}

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
💡 INSIGHT
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Overall media sentiment today is mostly
{'POSITIVE 🟢 - Good news dominates!' if len(positive) > len(negative) else 'NEGATIVE 🔴 - Challenging news cycle.' if len(negative) > len(positive) else 'NEUTRAL 🟡 - Balanced news today.'}

╚══════════════════════════════════════════════════════════╝
"""

print(report)

# Save report to file
with open("news_sentiment_report.txt", "w") as f:
    f.write(report)

print("✅ Report saved to news_sentiment_report.txt!")
```
